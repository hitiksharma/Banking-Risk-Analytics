We convert numerical data into categorical data

```python
bins = [0,100000, 300000, float('inf')]
labels = ['Low','Med','High']

df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins, labels=labels,right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique categories in categorical columns

categorical_cols = df[['BRId', 'GenderId', 'Amount of Credit Cards',
                       'Nationality', 'Occupation', 'Fee Structure',
                       'Loyalty Classification', 'Properties Owned',
                       'Risk Weighting', 'IAId', 'Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

#Univariate Analysis

```python
for i,predictor in enumerate(df[['BRId', 'GenderId', 'Amount of Credit Cards',
                       'Nationality', 'Occupation', 'Fee Structure',
                       'Loyalty Classification', 'Properties Owned',
                       'Risk Weighting', 'IAId', 'Income Band']]):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)
```

#Bivariate Analysis

```python
for i,predictor in enumerate(df[['BRId', 'GenderId', 'Amount of Credit Cards',
                       'Nationality', 'Occupation', 'Fee Structure',
                       'Loyalty Classification', 'Properties Owned',
                       'Risk Weighting', 'IAId', 'Income Band']]):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
#Hist plot of value conts of different occupations

for col in categorical_cols:
  if col == 'Occupation':
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("count")
  plt.show()
```

# Numerical Analysis


```python
# Numerical Analysis

num_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate analysis and visualisation

plt.figure(figsize=(15,15))
for i,col in enumerate(num_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

## Heatmaps

```python
correlation_matrix = df[num_cols].corr()

plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f')

plt.title("Correlation Matrix")
plt.show()
```

Bank deposits has high correlation with checking accounts and saving accounts and wrt to foreign currency accounts

this teels that the customers who maintain large balances in one account type often hold substantial amount of funds across other types as well

credit card balance (credit usage patterns vary indepedently of income level or loan amounts)
